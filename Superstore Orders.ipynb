# Top 5 Highest Selling Products in Each Region

```python
query2="""with xx as (select region,product_id,sum(quantity) as all_quantity from orders group by region,product_id)
select * from (select*, row_number() over(partition by region order by all_quantity desc) as rn from xx )x where rn<=5;"""

data2=pd.read_sql(query2,con=engine1)
data2
```

# Month-over-Month Growth Comparison (2022 vs 2023)

```python
query3="""with xx as (select year(order_date) oy,month(order_date) om ,sum(sale_price) as sales from orders group by oy,om)
select om,
sum(case when oy=2022 then sales else 0 end) as 2022_sales,
sum(case when oy=2023 then sales else 0 end) as 2023_sales
from xx
group by om order by om;"""

data3=pd.read_sql(query3,con=engine1)
data3
```

# Highest Sales Month for Each Category

```python
query4="""with xx as (select category,month(order_date) as ym,sum(sale_price)as sales from orders group by category,ym)
select * from (select *, row_number()over(partition by category order by sales desc)as rn from xx) x where rn=1"""

data4=pd.read_sql(query4,con=engine1)
data4
```

# Sub-Category with Highest Profit Growth in 2023 vs 2022

```python
query5="""with xx as (select sub_category,year(order_date) as yd,sum(profit) as total_profit from orders group by sub_category,yd)
, xx2 as(select sub_category,
sum(case when yd=2022 then total_profit else 0 end) as 2022_profit,
sum(case when yd=2023 then total_profit else 0 end) as 2023_profit
from xx group by sub_category)
select *,(2023_profit-2022_profit)*100/2022_profit from xx2 order by (2023_profit-2022_profit)*100/2022_profit desc;"""

data5=pd.read_sql(query5,con=engine1)
data5
```
